## 7 Scrapper Automatization

```python
def get_info(url):
    #diccionary for enter information
    dict = {}
    r = requests.get(url)
    page = BeautifulSoup(r.text,'lxml')
    
    #get title article
    title = page.find('h1', attrs={'class':'admania-entrytitle'})
    if title:
        dict['title'] = title.text
    else:
        dict['title'] = None
        
    #get_date_article 
    date = page.find('div', attrs={'class':'admania-entrydate'})
    if date:
        dict['date'] = date.text
    else:
        dict['date'] = None
    
    #get category
    category = page.find('div', attrs={'class':'admania-entrycategory'})
    if category:
        dict['category'] = category.a.get_text()
    else: 
        dict['category'] = None
    
    #get feauture image
    media = page.find('div', attrs={'class':'admania-single-post-ftimage'})
    if media:
        image = media.find_all('img')
        if len(image) == 0:
            print('there is not images')
        else:
            img = image[0]
            img_src = img.get('data-src')
            try:
                img_req = requests.get(img_src)
                if img_req.status_code == 200:
                    dict['img'] = img_req.content
                else:
                    dict['img'] = None
            except:
                print('There is not image')
    else:
        print('There is not media')
    
    return dict 
```

```python
url = 'https://www.filetechn.com'
def get_enter_links(url):
    try:
        r = requests.get(url)
        soup = BeautifulSoup(r.text,'lxml')
        titles = soup.find_all('h2', attrs={'class':'admania-entrytitle'})
        links_titles = [title.a.get('href') for title in titles]
    except Exception as e:
        print(e)
    
    return links_titles
```

```python
# GET LINKS ARTICLES IN 3 PAGES
list = ['https://www.filetechn.com/']
all_list = []

for i in range(2):
    r = requests.get(list[-1])
    soup = BeautifulSoup(r.text,'lxml')
    get_next_page = soup.find('a', attrs={'class':'next page-numbers'})
    get_link = get_next_page.get('href')
    list.append(get_link)
    
for i in list:
    all_list.extend(get_enter_links(i))
```

```python
# SCRAPER 
data = []
for i, link in enumerate(all_list):
    print(f'scraper_link{i}/{len(all_list)}')
    data.append(scraper_url(link))
```
