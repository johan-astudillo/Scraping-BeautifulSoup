# 4. Errors

```python
r = requests.get(links_titles[0])
if r.status_code == 200:
    # Success
    print('Success')
else:
    #Fail
    print('Fail')
```

```python
link_url = links_titles[0]
print(link_url)
```

```python
#create a bad url
bad_url = link_url.replace('t','a')
print(bad_url)
```

```python
# get a error because URL is not exists, then we use a try except
r = requests.get(bad_url)
if r.status_code == 200:
    # Success
    print('Success')
else:
    #Fail
    print('Fail')
```

```python
try:
    r = requests.get(bad_url)
except:
    print('Error request!\n')
    
print('Program continues...')
```

```python
try:
    r = requests.get(bad_url)
except Exception as e:
    print('Error request!\n')
    print(e)
    print('\n')
    
print('Program continues...')
```
